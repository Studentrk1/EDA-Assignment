# Check for missing values in all columns and describe your approach for handling them.(Include your Python code and output in the code box below.)

```python
import pandas as pd
import numpy as np

# Create a sample DataFrame with missing values
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 7, 8, 9, 10],
    'C': ['X', 'Y', 'Z', np.nan, 'W'],
    'D': [11, 12, 13, 14, 15]
}
df = pd.DataFrame(data)

# Check for missing values in all columns
missing_values = df.isnull().sum()

print("Missing values in each column:")
print(missing_values)

# Example of handling missing values (Imputation with mean for numerical, mode for categorical)
# Fill missing values in column 'A' with its mean
if missing_values['A'] > 0:
    df['A'].fillna(df['A'].mean(), inplace=True)

# Fill missing values in column 'B' with its median
if missing_values['B'] > 0:
    df['B'].fillna(df['B'].median(), inplace=True)

# Fill missing values in column 'C' with its mode
if missing_values['C'] > 0:
    df['C'].fillna(df['C'].mode()[0], inplace=True)

print("\nDataFrame after handling missing values:")
print(df)
```

# Plot the distribution of selling prices using a histogram and describe the overall trend.(Include your Python code and output in the code box below.)

# Compute the average km_driven for each ownership type (1st owner, 2nd owner, etc.), and present the result as a bar plot.(Include your Python code and output in the code box below.)

```python
import pandas as pd
import matplotlib.pyplot as plt
import io

# Create a sample DataFrame for demonstration purposes
data = {
    'ownership_type': ['1st owner', '2nd owner', '1st owner', '3rd owner', '2nd owner', '1st owner'],
    'km_driven': [50000, 75000, 60000, 90000, 80000, 55000]
}
df = pd.DataFrame(data)

# Calculate the average km_driven for each ownership type
average_km_driven = df.groupby('ownership_type')['km_driven'].mean().sort_values()

# Plot the results as a bar plot
plt.figure(figsize=(8, 6))
average_km_driven.plot(kind='bar', color='skyblue')
plt.title('Average Kilometers Driven by Ownership Type')
plt.xlabel('Ownership Type')
plt.ylabel('Average Kilometers Driven')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save the plot to a BytesIO object to capture the output
buffer = io.BytesIO()
plt.savefig(buffer, format='png')
buffer.seek(0)
plot_output = buffer.getvalue()

plt.show() # Display the plot

print("Average km_driven per ownership type:")
print(average_km_driven)
print("\nBar plot generated showing average km_driven per ownership type.")
```

# Use the IQR method to detect and remove outliers from the km_driven column. Show before-and-after summary statistics.(Include your Python code and output in the code box below.)

```python
import sklearn
from sklearn.datasets import load_diabetes
import pandas as pd
diabetes = load_diabetes()

column_name = diabetes.feature_names
df_diabetes = pd.DataFrame(diabetes.data)
df_diabetes .columns = column_name
df_diabetes .head()
print("Old Shape: ", df_diabetes.shape)


Q1 = df_diabetes['bmi'].quantile(0.25)
Q3 = df_diabetes['bmi'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

upper_array = np.where(df_diabetes['bmi'] >= upper)[0]
lower_array = np.where(df_diabetes['bmi'] <= lower)[0]

df_diabetes.drop(index=upper_array, inplace=True)
df_diabetes.drop(index=lower_array, inplace=True)

print("New Shape: ", df_diabetes.shape)
```

The most important factors affecting a bike's selling price are its condition, age, brand, model type, and mileage, with condition often being the most critical factor. Common data cleaning actions include handling missing values, removing inconsistent or duplicate entries, and correcting typographical errors. Feature engineering involves creating new, more informative features from existing ones, such as calculating the bike's age in years or grouping brands into categories.
* Factors Affecting Bike Selling Price
* Condition:
The overall physical state of the bike, including its frame, components, and mechanical function, is a major determinant of its value.
* Age:
Newer bikes generally command higher prices than older ones.
* Brand and Model:
Reputable brands and specific models (e.g., performance, utility) have different inherent values.
* Mileage/Usage:
Lower mileage typically indicates less wear and tear, increasing the bike's worth.
* Features and Customizations:
Aftermarket parts or specific upgrades can add value.
* Maintenance History:
Well-documented maintenance records can increase buyer confidence and the * selling price.
Data Cleaning and Feature Engineering Performed
* Data Cleaning:
Missing Value Imputation: Address any missing data points (e.g., mileage, condition descriptions) by either removing the affected rows or filling them with sensible values, such as the mean or median.
* Outlier Detection: Identify and handle any unusually high or low values for numerical features like price or mileage.
Standardization: Correct inconsistencies in categorical data, such as ensuring "Honda" and "honda" are treated as the same brand.
* Duplicate Removal: Remove any identical entries for the same bike to avoid distorting the dataset.
* Feature Engineering:
Age Calculation: Convert the manufacturing year into a more direct feature, such as "Bike Age" in years.
* Categorical Encoding: Transform text-based features like "brand" or "model type" into numerical representations (e.g., one-hot encoding) that machine learning models can understand.
* Binning: Group continuous features, like mileage or price, into discrete bins (e.g., "low mileage," "high mileage") if it provides better predictive power for the model.
* Interaction Features: Combine existing features to create new ones, such as a feature for "Brand_Condition" if certain brands consistently perform better in specific conditions.
